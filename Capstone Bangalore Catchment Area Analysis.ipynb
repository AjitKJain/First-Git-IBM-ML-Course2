Taking three variables i.e. shopping, eatout and amenities in separate dataset to build regression model.

```python
z3=z2[['Shopping','EatOut','Eminities']]
z3['Remark']='Keep'
for i in range(0,z3.shape[0]):
    if z3['Shopping'][i]<2:z3['Remark'][i]='Delete'
    if z3['EatOut'][i]<3:z3['Remark'][i]='Delete'
    if z3['Eminities'][i]<1:z3['Remark'][i]='Delete'

z3=z3[z3['Remark']=='Keep']
z3=z3.drop(['Remark'],axis=1)
```

As we saw data is very dispersed and any amount of transformation or feature scaling will not serve the purpose.
Taking median (with 10 percent tolerance) of Eatout and Aminities variable at each level of Shopping variable.

```python
y=pd.DataFrame()
for i in z3['Shopping'].unique():
    x=z3[z3['Shopping']==i]
    if x.shape[0]>=5:
        low=round(stat.median(x['EatOut'])*.9,0)
        high=round(stat.median(x['EatOut'])*1.1,0)
        x=x[x['EatOut']>=low]
        x=x[x['EatOut']<=high]
        y=y.append(x,ignore_index=True)
z3=y
y=pd.DataFrame()
for i in z3['Shopping'].unique():
    x=z3[z3['Shopping']==i]
    if x.shape[0]>=4:
        low=round(stat.median(x['Eminities'])*.9,0)
        high=round(stat.median(x['Eminities'])*1.1,0)
        x=x[x['Eminities']>=low]
        x=x[x['Eminities']<=high]
        y=y.append(x,ignore_index=True)

z3=y    
```

#### Fitting Simple Linear Regression

```python
x=np.array(z3[['EatOut','Eminities']])
y=np.array(z3['Shopping'])

reg=LinearRegression().fit(x,y)
print(reg.coef_)
print(reg.intercept_)
```

Using the model to predict number of shopping venues at each location (point in grid). Predicted values are in variable RegShops.

```python
z2['RegShops']=9999
for i in range(0,z2.shape[0]):
    z2['RegShops'][i]=reg.intercept_+reg.coef_[0]*z2['EatOut'][i]+reg.coef_[1]*z2['Eminities'][i]
```

Variable ShopDiff is created which is difference between predicted shopping venues (RegShops) and actual shopping venues.

Tagging locations as, if number of actual shopping venues is within twenty percent tolerance of predicted venues count then that location is saturated, if greater than twenty percent then competition is high, if actuals is les then twenty percent of predicted then there is opportunity.

If in any observation if any two variables among Shopping, EatOut and Amenities is zero then location is marked indecisive. Also tagging indecisive as per earlier rule that at any location there should be at least two shopping, three eatout and one amenity venue.

Next transforming variable similar to one hot encoding to prepare for DBSCAN clustering

```python
z2['ShopDiff']=z2['RegShops']-z2['Shopping']
z2['Remark']="Saturated"
for i in range(0,z2.shape[0]):
    if z2['Shopping'][i]>=round(z2['RegShops'][i]*1.2,0):z2['Remark'][i]='HighCompetition'
    if z2['Shopping'][i]<=round(z2['RegShops'][i]*.8,0):z2['Remark'][i]='Opportunity'
    x=[z2['Shopping'][i],z2['Eminities'][i],z2['EatOut'][i]]
    if x[0]<2: z2['Remark'][i]='InDecisive'
    if x[1]<1: z2['Remark'][i]='InDecisive'
    if x[2]<3: z2['Remark'][i]='InDecisive'
    for j in range(0,3): 
        if x[j]>0:
            x[j]=1
    if sum(x)<2:z2['Remark'][i]='InDecisive'

z4=z2[['lat','lon','ShopDiff','Remark']]
z4['newlat']=z4['lat']-min(z4['lat'])
z4['newlon']=z4['lon']-min(z4['lon'])
z4['InDecisive']=0
z4['Opportunity']=0
z4['Saturated']=0
z4['HighCompetition']=0

z4['InDecisive'][z4['Remark']=='InDecisive']=1
z4['Opportunity'][z4['Remark']=='Opportunity']=1
z4['Saturated'][z4['Remark']=='Saturated']=1
z4['HighCompetition'][z4['Remark']=='HighCompetition']=1
```

DBSCAN is performed to takeout outliers and be sure of the solution by setting min sample parameters to 5. Which means at least five points in any cluster should showup to be marked as Opportunity, High competition etc.

```python
x=np.array(z4[['ShopDiff','newlat', 'newlon', 'InDecisive','Opportunity', 'Saturated', 'HighCompetition']])
clust=DBSCAN(eps=.5,min_samples=5).fit(x)
z4['cluster']=clust.labels_
```
